```python
#exercicio 1 triplo de pitágoras
triplos_pitagoricos = set()  # Usando um set para armazenar triplos únicos

# Percorrendo todos os possíveis valores para lado1 e lado2
for lado1 in range(1, 1001):
    for lado2 in range(lado1, 1001):  # Começando de lado1 para evitar duplicatas
        quadrado_hipotenusa = lado1**2 + lado2**2
        hipotenusa = quadrado_hipotenusa ** 0.5

        # Verificando se a hipotenusa é um número inteiro e está dentro do limite
        if hipotenusa.is_integer() and hipotenusa <= 1000:
            # Ordenando os lados para evitar duplicatas devido à rotação dos catetos
            triplo = tuple(sorted([lado1, lado2, int(hipotenusa)]))
            triplos_pitagoricos.add(triplo)  # Adicionando o triplo ao conjunto

# Exibindo os triplos de Pitágoras únicos
for triplo in triplos_pitagoricos:
    print(triplo, end=' ')

#Um conjunto (set) é usado para armazenar os triplos de Pitágoras únicos encontrados.
#Dois loops são utilizados para iterar sobre os possíveis valores para lado1 e lado2.
#Para cada combinação de lado1 e lado2, é calculado o valor da hipotenusa e verificado se é um número inteiro dentro do limite de 1000.
#Os triplos são ordenados e adicionados ao conjunto para garantir a unicidade, considerando a rotação dos catetos.
#Por fim, os triplos de Pitágoras únicos são exibidos na saída.'''
```

```python
def leibniz_pi(num_terms):
    pi_approx = 0
    for i in range(num_terms):
        denominator = 2*i + 1
        if i % 2 == 0:
            pi_approx += 1/denominator
        else:
            pi_approx -= 1/denominator
    return 4 * pi_approx

# Função para arredondar o valor de π para a quinta casa decimal
def round_pi(value):
    return round(value, 5)

# Inicialização de variáveis
terms_to_check = 200000
found = False
terms_needed = 0

# Cálculo de π e verificação dos termos necessários
print("{:<10} {:<20}".format("Termo", "Valor de PI"))
for term in range(1, terms_to_check + 1):
    pi_value = round_pi(leibniz_pi(term))
    print("{:<10} {:<20}".format(term, pi_value))
    if str(pi_value)[:7] == '3.14159' and not found:
        found = True
        terms_needed = term

print("\nTotal de termos necessários para alcançar 3.14159: ", terms_needed)

```

```python
def is_magic_date(day, month, year):
    """
    Essa função verifica se uma data é mágica.

    Args:
    day (int): O dia da data.
    month (int): O mês da data.
    year (int): Os dois últimos dígitos do ano da data.

    Returns:
    bool: True se a data for mágica, False caso contrário.
    """
    return day * month == year

def find_magic_dates():
    """
    Esta função encontra e imprime todas as datas mágicas do século XX.
    """
    for year in range(1900, 2000):  # Loop pelos anos do século XX.
        for month in range(1, 13):  # Loop pelos 12 meses.
            for day in range(1, 32):  # Loop pelos dias. Considera até 31 dias, mesmo que alguns meses não tenham 31 dias.

                if is_magic_date(day, month, year % 100):  # Usa apenas os dois últimos dígitos do ano.
                    print(f"{day:02d}/{month:02d}/{year} é uma data mágica.")

# Chamada da função para encontrar e exibir todas as datas mágicas do século XX.
find_magic_dates()
```
